```python
data.drop('Unnamed: 32',inplace=True,axis=1)
```

```python
data['diagnosis']=pd.get_dummies(data['diagnosis'],dtype=int,drop_first=True) 
# Malignant for 1 and Benign for 0
```

```python
data['diagnosis'].hist() 
# we can observe that there is some class imbalance in the dataset that we can resolve or we can use different metrics for model 
# evaluation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(data.corr(),cmap='coolwarm',annot=True)
# using this we can remove the highly correlated columns from our data 
data.drop(['radius_mean','area_mean'])
```

```python
l=[]
corr_matrix=data.corr()
for i in corr_matrix.columns:
    for j in corr_matrix.columns:
        if corr_matrix[i][j] > 0.85 and i != j:
            if j not in l:
                l.append(j)  
# Out of totla 32 columns I am dropping 17 columns as they have very correlation with other columns, thereofre we don't need them
# anymore.
```

```python
data.drop(l,axis=1,inplace=True)
```

```python
data.drop('id',axis=1,inplace=True)
```

```python
y=data['diagnosis']
data.drop('diagnosis',axis=1,inplace=True)
```

```python
x=data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=102)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
model=lr.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,roc_auc_score
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
print(accuracy_score(y_test,predictions))
# We have a 71% of accuracy but we can't truust this as there is class imbalance in the dataset 
```

```python
y_pred_proba=model.predict_proba(X_test)[:,1]
```

```python
roc_auc_score(y_test,y_pred_proba) 
# SO the real accuracy score is 78% for our model 
```
